```python
X = location_pd.drop(['outcome', 'GHI', 'Latitude', 'Longitude'], axis = 1)
y = location_pd['outcome']
class_le = LabelEncoder()
y = class_le.fit_transform(y)
```

```python
stdc = StandardScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size = 0.3, random_state = 42, stratify=y)
```

```python
logistic = LogisticRegression(multi_class='multinomial', solver = 'lbfgs', penalty = 'l2', max_iter = 10000, random_state= 42)
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
```
